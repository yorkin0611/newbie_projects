Answer to the following Analytical questions: 

Question 1: Use Histograms to compare the disribution of movie `imdb_score` column related to movies produced in `France` and `Canada`. 

While creating the histgoram:
- Set the number of bins to 20 
- Set the xAxis Label 
- Set the yAxis Label
- Remove the grid lines


```python
# Create the histogram for Canada here 
ax = data[data.country =='Canada'].imdb_score.hist(bins=20, grid=False)
ax.set_title('IMDB Scores for Canada productions')
ax.set_xlabel('IMDB Scores')
ax.set_ylabel('Counts')
```

```python
# Create the histogram for France here 
ax = data[data.country =='France'].imdb_score.hist(bins=20, grid = False)
ax.set_title('IMDB Scores for France productions')
ax.set_xlabel('IMDB Scores')
ax.set_ylabel('Counts')
```

Question 2: Create a histogram to check the distribution of `budget` spent to produce Black&White movies.

While creating the histgoram:
- Set the number of bins to 10 
- Set the xAxis Label 
- Set the yAxis Label
- Remove the grid lines


```python
ax = data[data.color=='Black and White'].budget.hist(bins=10, grid=False)
ax.set_title('B&W Movie Budgets')
ax.set_xlabel('Budgets')
ax.set_ylabel('Counts')
```
