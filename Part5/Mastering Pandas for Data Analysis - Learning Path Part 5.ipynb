# TASK 6. PERFORM TEXT DATA CLEANING BY REMOVING PUNCTUATIONS

```python
# String module is super useful when dealing with text data
# String contains constants and classes for working with text
import string
string.punctuation
```

```python
Test = '$I love Pandas &Data Analytics!!'
```

```python
Test_punc_removed = [char for char in str(Test) if char not in string.punctuation]
#Test_punc_removed = []
#for char in str(Test):
#    if char not in string.punctuation:
#        Test_punc_removed.append(char)

Test_punc_removed
```

```python
# Join the characters again to form the string.
Test_punc_removed_join = ''.join(Test_punc_removed)
Test_punc_removed_join
```

```python
# Let's define a function to remove punctuations
def remove_punc(message):
    #Test_punc_removed = []
    #for char in str(message):
    #    if char not in string.punctuation:
    #        Test_punc_removed.append(char)
    Test_punc_removed = [char for char in str(message) if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)

    return Test_punc_removed_join
```

```python
# Let's remove punctuations from our dataset
df['reviews_nopunc']= df['verified_reviews'].apply(remove_punc)
df
```

# TASK 7. PERFORM TEXT DATA CLEANING BY REMOVING STOPWORDS

```python
!pip install gensim -q
!pip install nltk -q
print("installed gensim and nltk")
```

```python
import nltk
from nltk.corpus import stopwords
import gensim
from gensim.utils import simple_preprocess
```

```python
# download stopwords, ie those that often appear but don't add any value in ML
nltk.download("stopwords")
stopwords.words('english')
```

```python
# Add more stopwords by using extend
stop_words = stopwords.words('english')
stop_words.extend(['Amazon', 'amazon', 'alexa', 'echo', 'Alexa','device', 'Dot', 'dot'])
```

```python
# Simple_preprocess converts a string into a series of lowered case tokens
# Let's try it on a sample dataset
gensim.utils.simple_preprocess(df['verified_reviews'][0])
#returns the list of cleaned up words
```

```python
# Remove stopwords and remove short words (less than 2 characters)
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if len(token) >= 1 and token not in stop_words:
            result.append(token)

    return result
```

```python
# apply pre-processing to the text column
df['reviews_nopunc_nostopwords'] = df['reviews_nopunc'].apply(preprocess)
```
